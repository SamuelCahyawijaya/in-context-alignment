```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('icl_results.csv')
```

```python
df.model_name.unique()
```

```python
model = 'bloom-7b1'
for dset in df['dataset'].unique():
    if dset != 'masakhanews':
        continue
    pdf = df.loc[
        (df['dataset'] == dset) & (df['model_name'] == model) &
        (df['icl_idx'].isin(['sbert', 'random'])) & 
        (df['iia_idx'].isin(['sbert', 'random'])),:
    ].copy()
    
    for lang in pdf['lang'].unique():
        baseline = pdf.loc[
            (pdf['lang'] == lang) & (pdf['icl_type'] == 'none') & 
            (pdf['iia_type'] == 'none') & 
            (pdf['use_ioa'] == 'False'),['accuracy', 'macro_f1', 'weighted_f1']
        ].values
        
        # display(pdf.loc[pdf['lang'] == lang,:].sort_values(['dataset', 'icl_type', 'iia_type', 'use_ioa'], ascending=False).reset_index(drop=True).groupby(
        #     ['dataset', 'icl_type', 'iia_type', 'icl_idx', 'iia_idx', 'num_icl', 'num_iia', 'use_ioa']
        # )[['accuracy', 'macro_f1', 'weighted_f1']].mean() - baseline)
        
        pdf.loc[pdf['lang'] == lang,['accuracy', 'macro_f1', 'weighted_f1']] -= baseline
    
    display(pdf.sort_values(['dataset', 'icl_type', 'iia_type', 'use_ioa'], ascending=False).reset_index(drop=True).groupby(
        ['dataset', 'icl_type', 'iia_type', 'icl_idx', 'iia_idx', 'num_icl', 'num_iia', 'use_ioa']
    )[['accuracy', 'macro_f1', 'weighted_f1']].mean())
```
